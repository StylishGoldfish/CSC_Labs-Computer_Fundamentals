```python
from random import randint as computerRandom

playFlag = True
com_num = computerRandom(0,100)
score = 5 
#Initate Game - 
while playFlag == True and score != 0:
    play_num = int(input("Enter guess: ")) #Find player guess
    if play_num > com_num:
        print("Too high!")
        score -= 1
    elif play_num == com_num:
        print("Correct")
        playFlag = False
    elif play_num < com_num:
        print("Too low!")
        score -= 1
if score == 0:
    print("you lost")

```
