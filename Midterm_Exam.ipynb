```python
#Question 5 for Programming Section

for i in range(10, 50, 4):
    print(i*i)

# a) The number 10 in the range function is the starting value to be used in the for loop, the number 10 is included in the range
# b) The number 50 in the range function is the ending value to be used in the for loop, the number 50 is not included in the range
# c) The number 4 in the range function is the step value. It signifies the value at which each number created by the range function is incrementing by. 
# d) This program will print "100 196 324 484 676 900 1156 1444 1764 2116" where each number is printed on a new line
```

```python
# Question 6 for Programming Section

a = 87
b = a%7
print(b)
c = a/7
print(c)

# The code above will print '3 12.428571428571429' with both numbers appearing on a new line
# The reason why, is because first the variable 'a' has the int 87 in it. Then we mod 'a' with 7
# where we get 3(because it is the remainder)as our answer which is then stored into variable 'b'
# The second output is 'c' where we stored the floating point value a divided by 7 which calculates to 12.428571428571429
```
