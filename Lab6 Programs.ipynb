```python
SCORES = [40,91,85,15]
n = len(SCORES)
newList=[]
    # Traverse through all array elements
for i in range(n):
    count = 1
    for j in range(n):
        try: 
            pair=[SCORES[i], SCORES[i+count]]
            newList.append(pair)
            count+=1
        except IndexError: 
            break

for i in newList:
    print(i)
```
