```python
# EVEN OR ODD PROGRAM

userInput = 0
evenChecker = 2

userInput = int(input("Enter a number: "))

if userInput%evenChecker == 0: 
    print("even number")
    if userInput<0:
        print("negative number")
else: 
    print("odd number")
    if userInput<0:
        print("negative number")


```

```python
# Variable initializing
iteration = 0 # Counts what iteration of while loop we are on, and what number we're checking
evenCount = 0 # Couts how many even numbers have been found till now

while evenCount < 20:   # Loops until we've found 20 even numbers
    if iteration%2 == 0:  
        print(iteration)
        iteration += 1
        evenCount += 1
    else:
        iteration += 1
```
