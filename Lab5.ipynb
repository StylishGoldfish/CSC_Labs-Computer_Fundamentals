Find the length of all the strings in the below list. Use a for loop or a while loop. (10 points) 

```
names = [“Jennifer”, “Albatross”, “Justin”, “Dave”, “Shankarnarayan”, “Ezra”, “Alice”, “Kwabena”]
```

````python
names = ["Jennifer", "Albatross", "Justin", "Dave", "Shankarnarayan", "Ezra", "Alice", "Kwabena"]
letterCountList = []
for name in names:
    letterCounter = 0
    for letter in name:
        letterCounter += 1
    letterCountList.append(letterCounter)


for i in range(len(names)):
    print(names[i]+":", letterCountList[i])
````

Implement a Leaderboard using a List (20 points)


````python
import random 

top_scores= [0,0,0,0,0]

for i in range(50):
    rand_num = random.randint(0,100)

#This algorithm can self sort the top scores from highest to smallest
    for placeNumber,score in enumerate(top_scores):
        if rand_num > score:
            top_scores[placeNumber] = rand_num
            break
        else:
            continue

print("The Top Scores Of This Round Are:")
for i in range(len(top_scores)):
    print(str(i+1)+"th Place:", top_scores[i])
````

1.	What is bubblesort? Write a program that implements the bubblesort algorithm.	(10 points)

````python
def orderLeastToGreatest(listArgument):
    n = len(listArgument)
    for i in range(n):
        for j in range(0, n-i-1):
            if listArgument[j]>listArgument[j+1]:
                listArgument[j], listArgument[j+1] = listArgument[j+1], listArgument[j]
            else:
                continue
    return listArgument

    
listArgument = [12,23,45,34,22,12,45]
orderLeastToGreatest(listArgument)
print(listArgument)
````
